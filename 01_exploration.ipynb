```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

pd.set_option("display.max_columns", None)

data_dir = Path("data")
csv_files = list(data_dir.glob("*.csv"))

print(f"Found {len(csv_files)} CSV files")
csv_files[:5]
```

```python
dfs = []

for file in csv_files:
    temp = pd.read_csv(file, low_memory=False)
    temp["source_file"] = file.name
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True)

df.shape
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
df.columns.tolist()
```

```python
required_cols = [
    "ClosePrice",
    "LivingArea",
    "Bedrooms",
    "Bathrooms",
    "LotSize",
    "PropertyType",
    "PropertySubType"
]

for col in required_cols:
    print(col, col in df.columns)
```

```python
df_filtered = df[
    (df["PropertyType"] == "Residential") &
    (df["PropertySubType"] == "SingleFamilyResidence")
].copy()

df_filtered.shape
```

```python
df_filtered["PropertyType"].value_counts(dropna=False)
```

```python
df_filtered["PropertySubType"].value_counts(dropna=False)
```

```python
[col for col in df_filtered.columns if "date" in col.lower()]
```

```python
df_filtered["CloseDate"] = pd.to_datetime(df_filtered["CloseDate"], errors="coerce")

df_filtered["CloseDate"].min(), df_filtered["CloseDate"].max()
```

```python
date_min = df_filtered["CloseDate"].min()
date_max = df_filtered["CloseDate"].max()

print("Earliest date:", date_min)
print("Latest date:", date_max)
print("Days covered:", (date_max - date_min).days)
print("At least 6 months?", (date_max - date_min).days >= 180)
```

```python
for keyword in ["Close", "Living", "Bedroom", "Bathroom", "LotSize", "PropertyType", "PropertySubType"]:
    print(f"\nColumns containing '{keyword}':")
    print([col for col in df_filtered.columns if keyword.lower() in col.lower()])
```

```python
df_filtered = df_filtered.rename(columns={
    "BedroomsTotal": "Bedrooms",
    "BathroomsTotalInteger": "Bathrooms",
    "LotSizeSquareFeet": "LotSize"
})
```

```python
eda_cols = [
    "CloseDate",
    "ClosePrice",
    "LivingArea",
    "Bedrooms",
    "Bathrooms",
    "LotSize",
    "PropertyType",
    "PropertySubType",
    "source_file"
]

df_eda = df_filtered[eda_cols].copy()

df_eda.head()
```

```python
df_eda["CloseDate"] = pd.to_datetime(df_eda["CloseDate"], errors="coerce")

numeric_cols = ["ClosePrice", "LivingArea", "Bedrooms", "Bathrooms", "LotSize"]

for col in numeric_cols:
    df_eda[col] = (
        df_eda[col]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    df_eda[col] = pd.to_numeric(df_eda[col], errors="coerce")
```

```python
print("Rows after filtering:", df_eda.shape[0])
print("Columns:", df_eda.shape[1])

print("\nPropertyType values:")
print(df_eda["PropertyType"].value_counts(dropna=False))

print("\nPropertySubType values:")
print(df_eda["PropertySubType"].value_counts(dropna=False))

print("\nEarliest CloseDate:", df_eda["CloseDate"].min())
print("Latest CloseDate:", df_eda["CloseDate"].max())
print("Days covered:", (df_eda["CloseDate"].max() - df_eda["CloseDate"].min()).days)
print("At least 6 months?", (df_eda["CloseDate"].max() - df_eda["CloseDate"].min()).days >= 180)
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(df_eda[col].dropna(), bins=50)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in numeric_cols:
    lower = df_eda[col].quantile(0.01)
    upper = df_eda[col].quantile(0.99)

    trimmed = df_eda[(df_eda[col] >= lower) & (df_eda[col] <= upper)]

    plt.figure(figsize=(8, 5))
    plt.hist(trimmed[col].dropna(), bins=50)
    plt.title(f"Distribution of {col}, Trimmed 1st-99th Percentile")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 3))
    plt.boxplot(df_eda[col].dropna(), vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
feature_cols = ["LivingArea", "Bedrooms", "Bathrooms", "LotSize"]

for col in feature_cols:
    plt.figure(figsize=(8, 5))
    plt.scatter(df_eda[col], df_eda["ClosePrice"], alpha=0.3)
    plt.title(f"ClosePrice vs {col}")
    plt.xlabel(col)
    plt.ylabel("ClosePrice")
    plt.show()
```

```python
df_eda.to_csv("residential_single_family_eda.csv", index=False)
```

## EDA Summary

I loaded the monthly Trestle property CSV files into pandas and combined them into one dataframe. I then restricted the dataset to records where `PropertyType = Residential` and `PropertySubType = SingleFamilyResidence`, as required by the task.

The analysis uses `ClosePrice` as the target variable, along with `LivingArea`, `BedroomsTotal`, `BathroomsTotalInteger`, and `LotSizeSquareFeet`. These were renamed to `LivingArea`, `Bedrooms`, `Bathrooms`, and `LotSize` for readability.

The filtered dataset covers approximately six months of closed residential single-family listings. I explored the distributions of `ClosePrice`, `LivingArea`, `Bedrooms`, `Bathrooms`, and `LotSize` using histograms and boxplots. I also created scatterplots comparing each feature against `ClosePrice`.

Initial observations:
- `ClosePrice` is right-skewed, with many homes in a lower-to-middle price range and some high-price outliers.
- `LivingArea` and `LotSize` are also right-skewed, which is common in real estate data.
- `Bedrooms` and `Bathrooms` are discrete variables, concentrated around common household counts.
- `LivingArea` appears positively related to `ClosePrice`, meaning larger homes generally tend to sell for more.
- `LotSize` contains large outliers, so trimmed plots are useful for visualization.
